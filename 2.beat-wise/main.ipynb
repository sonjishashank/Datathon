### using another method i.e for loop for checking null values

```python
features= ['city', 'beat']
```

```python
for feature in features:
    null_check = pd.isnull(df[feature])
    print(df[null_check])
```

### droping the rows with null values

```python
df = df.dropna()
df.shape
```

```python
df['beat'].unique()
```

### Grouping the crimes beat wise

```python
df_crime = df.groupby('beat').size().reset_index(name = 'Numbers')
df_crime
```

### Plotting the graph for analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(210,8))
plt.xlabel('Beat')
plt.title('Bangalore Beat Wise Crime Distribution')
plt.ylabel('Number of crimes')

x = df_crime['beat']
y = df_crime['Numbers']

# Plotting with adjusted x-axis labels
plt.bar(x, y)
plt.xticks(rotation=90)  # Rotate x-axis labels by 45 degrees for better readability

plt.grid(which='both', linestyle=':')
plt.tight_layout()

# Save the plot with adjusted x-axis labels
plt.savefig('new2.jpg', dpi=300)
```
