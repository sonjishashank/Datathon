```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
crime_data = pd.read_csv('crime_data.csv')  # Replace 'crime_data.csv' with your file path

# Preprocessing (if needed)
# For example, convert date/time columns to datetime format

# Group data by beat and count the number of crimes in each beat
beat_crime_counts = crime_data['Beat'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
beat_crime_counts.plot(kind='bar')
plt.title('Beat-wise Distribution of Crimes')
plt.xlabel('Beat')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
